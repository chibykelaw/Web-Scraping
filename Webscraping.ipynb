## Import Libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

## Send Requests

```python
url = "https://www.jumia.com.ng/computing/laptops/"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
#Product Name

product_name = []

product_name_item = soup.find_all("h3", {"class":"name"})

for item in product_name_item:
    product_name.append(item.text)
```

```python
# PAroduct Price

product_price = []
product_price_item = soup.find_all("div", {"class":"prc"})

for item in product_price_item:
    product_price.append(item.text)
```

## Create a Pandas Dataframe

```python
final_array = []

for name, price in zip(product_name, product_price):
    final_array.append({"Product Name":name, "Product Price":price})
    

df = pd.DataFrame(final_array)
df
```

```python
 df.to_csv('Jumia Laptops Prices')
```
